```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Seleccionar imagen al entorno

```python
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

# Definir filtros (kernels)

```python
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])  # Filtro Sobel en dirección X

sobel_y = np.array([[-1, -2, -1],
                    [ 0,  0,  0],
                    [ 1,  2,  1]])  # Filtro Sobel en dirección Y

laplacian = np.array([[ 0,  1,  0],
                       [ 1, -4,  1],
                       [ 0,  1,  0]])  # Filtro Laplaciano
```

# Aplicar convolución con los filtros

```python
sobel_x_result = cv2.filter2D(img, -1, sobel_x)
sobel_y_result = cv2.filter2D(img, -1, sobel_y)
laplacian_result = cv2.filter2D(img, -1, laplacian)
```

# Mostrar resultados

```python
fig, axes = plt.subplots(1, 4, figsize=(15,5))
axes[0].imshow(img, cmap='gray')
axes[0].set_title("Imagen Original")
axes[1].imshow(sobel_x_result, cmap='gray')
axes[1].set_title("Filtro Sobel X")
axes[2].imshow(sobel_y_result, cmap='gray')
axes[2].set_title("Filtro Sobel Y")
axes[3].imshow(laplacian_result, cmap='gray')
axes[3].set_title("Filtro Laplaciano")

for ax in axes:
    ax.axis("off")

plt.show()
```
